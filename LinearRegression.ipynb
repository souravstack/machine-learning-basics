#
Lost and cost function



```python
df['Predicted_y'] = reg.predict(x) # Assigning the predicted values to the 'Predicted_y' column in the DataFrame.
df['loss'] = df['Ending (USD)'] - df['Predicted_y'] # Calculate the 'loss' column based on 'Ending (USD)' and 'Predicted_y' column.
df.head() # Displaying the first few rows of the DataFrame.
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(df['Ending (USD)'],df['Predicted_y'])
mse
```

```python
mae=mean_absolute_error(df['Ending (USD)'],df['Predicted_y'])
mae
```
